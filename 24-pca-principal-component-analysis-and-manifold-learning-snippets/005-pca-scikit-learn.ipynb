```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

```python
# Making a DataFrame to view data easily
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
# help(PCA)
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_X)
```

```python
pca.transform(scaled_X)
```

```python
pc_results = pca.transform(scaled_X)
```

```python
plt.figure(figsize=(12, 6), dpi=200)
plt.scatter(pc_results[:, 0], pc_results[:, 1])
```

```python
plt.figure(figsize=(12, 6), dpi=200)
plt.scatter(pc_results[:, 0], pc_results[:, 1], c=cancer['target'])
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, index=['PC1', 'PC2'], columns=df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20, 3), dpi=200)
sns.heatmap(df_comp, annot=True)
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30.fit(scaled_X)
```

```python
pca_30.explained_variance_ratio_
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1, 30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1, 30), explained_variance)

plt.xlabel('Num of components')
plt.ylabel('Variance Explained')
```
