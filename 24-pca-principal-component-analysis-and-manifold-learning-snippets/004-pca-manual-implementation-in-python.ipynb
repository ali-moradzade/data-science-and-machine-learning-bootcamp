```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
# Making a DataFrame to view data easily
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
sns.heatmap(data=df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean()
```

```python
scaled_X.mean(axis=0)
```

```python
scaled_X - scaled_X.mean(axis=0)
```

```python
scaled_X
```

```python
covariance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors.shape
```

PCA 30 Feature Space --> N PC space --> 2

How many PC to choose? N = 2

```python
num_components = 2
```

```python
np.argsort([2, 1, 3])
```

```python
np.argsort(eigen_values)
```

```python
np.argsort(eigen_values)[::-1]
```

```python
np.argsort(eigen_values)[::-1][:num_components]
```

```python
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

```python
eigen_values, eigen_vectors = eigen_values[sorted_key], eigen_vectors[:,sorted_key]
```

```python
principal_components = np.dot(scaled_X, eigen_vectors)
```

```python
principal_components
```

```python
plt.figure(figsize=(12, 6), dpi=200)
plt.scatter(principal_components[:,0], principal_components[:,1])
```
