```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cluster_mpg.csv')
```

```python
df_w_dummies = pd.get_dummies(df.drop('name', axis=1))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_w_dummies)
```

```python
scaled_df = pd.DataFrame(data=scaled_data, columns=df_w_dummies.columns)
```

```python
plt.figure(figsize=(12, 6), dpi=200)

sns.clustermap(data=scaled_df, col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
scaled_df.head()
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12, 6), dpi=200)
sns.scatterplot(data=df, x='mpg', y='weight')
```

```python
plt.figure(figsize=(12, 6), dpi=200)
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_labels)
```

```python
plt.figure(figsize=(12, 6), dpi=200)
sns.scatterplot(data=df, x='mpg', y='horsepower', hue=cluster_labels,
               palette='viridis')
```

```python
from scipy.cluster import hierarchy
```

```python
linkage_metrix = hierarchy.linkage(model.children_)
```

```python
linkage_metrix
```

```python
plt.figure(figsize=(12, 8), dpi=200)

dendro = hierarchy.dendrogram(linkage_metrix)
```

```python
plt.figure(figsize=(12, 8), dpi=200)

dendro = hierarchy.dendrogram(linkage_metrix, truncate_mode='lastp', p=10)
```

```python
plt.figure(figsize=(12, 8), dpi=200)

dendro = hierarchy.dendrogram(linkage_metrix, truncate_mode='level', p=3)
```
