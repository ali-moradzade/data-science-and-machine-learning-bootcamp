```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='blue')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='lightblue')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='#650dbd') # RGB hex code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='#e6c017') # RGB hex code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='#e6c017', label='x') # RGB hex code
ax.plot(x, x + 1, color='purple', label='x + 1')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='#e6c017', lw=0.5) # RGB hex code
ax.plot(x, x + 1, color='purple', lw=5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='#e6c017', lw=1.5, ls='--') # RGB hex code
ax.plot(x, x + 1, color='purple', lw=5, linestyle='-.')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x, x + 1, color='purple', lw=5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x, x + 1, color='purple', lw=5)
lines[0].set_dashes([5,2,10,2]) # width of dash, with of space, width of dash, ...
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x, x + 1, color='purple', lw=5)
lines[0].set_dashes([5,2,10,2]) # width of dash, with of space, width of dash, ...
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x, x + 1, color='purple', lw=5)
lines[0].set_dashes([1,2,1,2,10,2])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+1, color='blue', lw=2, marker='o')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x+1, color='blue', lw=2, marker='o', ls='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='red', lw=2, marker='+', ls='--', ms=20)
ax.plot(x, x+1, color='purple', lw=2, marker='o', ls='--', ms=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x, x, color='red', lw=2, marker='+', ls='--', ms=20)
ax.plot(x, x+1, color='purple', lw=2, marker='s', ls='--', ms=10,
       markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
```
