```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of Four')

axes1.plot(a, b);

axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])

axes2.set_xlim(0, 8)
axes2.set_xlim(0, 100)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

axes2.plot(a, b);
```
