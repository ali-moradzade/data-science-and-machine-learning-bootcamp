```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../08-linear-regression-models/Advertising.csv')
```

```python
df
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df
```

```python
sns.scatterplot(data=df, x='total_spend', y='sales')
```

```python
sns.regplot(data=df, x='total_spend', y='sales')
```

```python
X = df['total_spend']
y = df['sales']
```

```python
# y = mx + b
# y = B1x + B0

# help(np.polyfit)
```

```python
np.polyfit(X, y, deg=1)
```

```python
B1, B0 = np.polyfit(X, y, 1)
```

```python
potential_spend = np.linspace(0, 500, 100)
```

```python
predicted_sales = B1 * potential_spend + B0
```

```python
plt.plot(potential_spend, predicted_sales)
```

```python
sns.scatterplot(x='total_spend', y='sales', data=df)
plt.plot(potential_spend, predicted_sales, c='red')
```

```python
sns.regplot(x='total_spend', y='sales', data=df)
```

```python
# y = B3y^3 + B2x^2 + B1x + B0
B3, B2, B1, B0 = np.polyfit(X,y,3)
```

```python
pot_spend = np.linspace(0, 500, 100)
```

```python
pred_sales =  B3 * pot_spend ** 3 + B2 * pot_spend ** 2 + B1 * pot_spend + B0
```

```python
plt.plot(pot_spend, pred_sales)
```

```python
sns.scatterplot(x='total_spend', y='sales', data=df)
plt.plot(pot_spend, pred_sales, c='r')
```
